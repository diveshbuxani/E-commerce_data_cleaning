```python
# Deleting redundant columns
df = df.drop(columns = "Unnamed: 0")
df.head()
```

```python
df = df.rename(columns = {"Deatils" : "Details"})
df.head()
```

```python
df.loc[df.duplicated()]
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Fixing the index since rows were deleted
df = df.reset_index().drop(columns = "index")
```

```python
# Removing NA values
df.isna().sum
df = df.dropna()
```

```python
df["BrandName"] = df["BrandName"].apply(lambda x : x.capitalize())
df["Details"] = df["Details"].apply(lambda x : x.capitalize())
```

```python
df["Sizes"].loc[27926]
# type(df["Sizes"].loc[27926])
# need to remove the Nan values
```

```python
df
```

```python
df["Sizes"] = df["Sizes"].apply(lambda x : "NA VALUED" if x == "Nan" else x)
```

```python
df
```

```python
df[df["Sizes"] != "NA VALUED"]
```

```python

for column in df.columns:
    rows_to_keep = (df[column] != 'Nan') & (df[column] != 'NA VALUED')
    df = df[rows_to_keep]

df
```

```python
df = df.reset_index().drop(columns = "index")
df
```

```python
df.to_csv("Cleaned_FashionDataset.csv", index = False)
```
